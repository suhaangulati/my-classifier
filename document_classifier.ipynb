# AG News Text Classification 📰

A simple, end-to-end NLP pipeline that classifies news headlines into four categories — **World**, **Sports**, **Business**, and **Sci/Tech** — using TF-IDF features and a Logistic Regression classifier.

## 1. Imports

```python
from datasets import load_dataset
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, accuracy_score
import joblib
```

## 2. Load data

Downloads the [AG News dataset](https://huggingface.co/datasets/fancyzhx/ag_news) from Hugging Face automatically the first time you run this cell.

```python
dataset = load_dataset('fancyzhx/ag_news')

LABEL_NAMES = ['World', 'Sports', 'Business', 'Sci/Tech']

X_train = dataset['train']['text']
y_train = dataset['train']['label']
X_test = dataset['test']['text']
y_test = dataset['test']['label']

print(f'Train examples: {len(X_train)}, Test examples: {len(X_test)}')
print('Categories:', LABEL_NAMES)
```

## 3. Build pipeline (TF-IDF + Logistic Regression)

```python
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(lowercase=True, stop_words='english', ngram_range=(1, 2))),
    ('clf', LogisticRegression(max_iter=1000)),
])
```

## 4. Train

```python
pipeline.fit(X_train, y_train)
```

## 5. Evaluate

```python
y_pred = pipeline.predict(X_test)
print('Accuracy:', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=LABEL_NAMES))
```

## 6. Predict on new, unseen headlines

```python
new_headlines = [
    'The central bank raised interest rates for the third time this year.',
    'The home team clinched the championship in a thrilling overtime finish.',
    'Scientists unveiled a new chip that doubles processing speed.',
]

predictions = pipeline.predict(new_headlines)
for text, pred in zip(new_headlines, predictions):
    print(f'{LABEL_NAMES[pred]:10s} <- {text}')
```

## 7. Save the trained model

```python
joblib.dump({'pipeline': pipeline, 'categories': LABEL_NAMES}, 'model.joblib')
print('Saved to model.joblib')
```
